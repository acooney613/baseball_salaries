```python
field = pd.read_csv('field.csv')
salary = pd.read_csv('salary.csv')
stats = pd.read_csv('stats.csv')

def batter_direction(name):
    if '*' in name:
        return 'L'
    elif '#' in name:
        return 'S'
    else:
        return 'R'
```
